```
def sigmoid(z):
    return 1/(1+np.exp(-z))

def loss(x_train, theta, y, _lambda=0):
    h = sigmoid(np.dot(x_train, theta))
    return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean() + (_lambda/y.shape[0])*np.sum(theta**2)

def gradient(x, theta, y, _lambda = 0):
    h = sigmoid(np.dot(x_train, theta))
    return (np.dot(x.T, (h - y)) + _lambda*theta) / y.shape[0]

def predict(x, theta, thresh=0.5):
    return sigmoid(np.dot(x, theta)) >= thresh
```

```
#loading data
train_data = np.load('train-ubyte.npz')
test_data  = np.load('test-ubyte.npz')

x_train, y_train = train_data['a'], train_data['b']
x_test,  y_test  = test_data['a'],  test_data['b']
x_train = x_train.reshape(x_train.shape[0],-1)
x_test = x_test.reshape(x_test.shape[0],-1)
```

```
theta = np.zeros([10, 784])
for i in range(10):
    clf = LogisticRegression(solver='lbfgs').fit(x_train, (y_train == i) * 1)
    theta[i,:] = clf.coef_
    y_true = (y_train == i)*1
    y_pred = clf.predict(x_train)
    tp = np.sum(np.logical_and(y_true == 1, y_pred == 1))
    tn = np.sum(np.logical_and(y_true == 0, y_pred == 0))
    fp = np.sum(np.logical_and(y_true == 0, y_pred == 1))
    fn = np.sum(np.logical_and(y_true == 1, y_pred == 0))
    cnf_matrix = [[tn, fp], [fn, tp]]
    train_loss = (fp+fn)/np.sum(cnf_matrix)
    print('Training loss is: ' + str(train_loss))

    y_true = (y_test == i)*1
    y_pred = clf.predict(x_test)
    tp = np.sum(np.logical_and(y_true == 1, y_pred == 1))
    tn = np.sum(np.logical_and(y_true == 0, y_pred == 0))
    fp = np.sum(np.logical_and(y_true == 0, y_pred == 1))
    fn = np.sum(np.logical_and(y_true == 1, y_pred == 0))
    cnf_matrix = [[tn, fp], [fn, tp]]
    test_loss = (fp+fn)/np.sum(cnf_matrix)
    print('Testing loss is: ' + str(test_loss))
    print('Confusion matrix is: ' + str(cnf_matrix))
```

```
h = sigmoid(np.dot(x_train, theta.T))
y_pred = np.argmax(h, axis=1)
cnf_matrix = confusion_matrix(y_train, y_pred)
print('Training confusion matrix is: \n' + str(cnf_matrix))
losses = np.sum(y_train != y_pred)/y_train.shape[0]
print('Training loss is: ' + str(losses))

h = sigmoid(np.dot(x_test, theta.T))
y_pred = np.argmax(h, axis=1)
cnf_matrix = confusion_matrix(y_test, y_pred)
print('Testing confusion matrix is: \n' + str(cnf_matrix))
losses = np.sum(y_test != y_pred)/y_pred.shape[0]
print('Testing loss is: ' + str(losses))
```
